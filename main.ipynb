# # Cluster Analysis

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import numpy as np

# Data Preprocessing
# Selecting numerical features for clustering
numerical_features = data.select_dtypes(include=[np.number]).columns.tolist()
clustering_data = data[numerical_features]

# Standardizing the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(clustering_data)

# Choosing the Number of Clusters using the Elbow Method
inertia = []
k_values = range(1, 11)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)

# Plotting the Elbow Method graph
plt.figure(figsize=(10, 6))
plt.plot(k_values, inertia, marker='o')
plt.title('Elbow Method to Determine Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.xticks(k_values)
plt.grid(True)
plt.show()
```

```python
from sklearn.decomposition import PCA

# Principal Component Analysis (PCA)
# Choosing the number of components such that 95% of the variance is retained
pca = PCA(n_components=0.95)
pca_data = pca.fit_transform(scaled_data)

# Explained variance ratio by the selected components
explained_variance = pca.explained_variance_ratio_

# Number of components chosen by PCA
n_components = pca.n_components_

# Performing K-means clustering on PCA-transformed data
# Using a reasonable number of clusters (e.g., 5) for initial exploration
kmeans_pca = KMeans(n_clusters=5, random_state=0)
kmeans_pca.fit(pca_data)
clusters = kmeans_pca.labels_

# Adding cluster information to the original data for analysis
data['Cluster'] = clusters

# Plotting the explained variance
plt.figure(figsize=(10, 6))
plt.bar(range(1, n_components + 1), explained_variance, alpha=0.7, align='center')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance by PCA Components')
plt.show()

n_components, explained_variance, data['Cluster'].value_counts()
```

```python
# Sampling the data (e.g., 10% of the data)
sampled_data = data.sample(frac=0.1, random_state=0)

# Standardizing the sampled data
scaled_sampled_data = scaler.transform(sampled_data[numerical_features])

# Applying PCA on the sampled data
pca_sampled = PCA(n_components=0.95)
pca_sampled_data = pca_sampled.fit_transform(scaled_sampled_data)

# Number of components and explained variance in the sampled data
n_components_sampled = pca_sampled.n_components_
explained_variance_sampled = pca_sampled.explained_variance_ratio_

# Applying K-means clustering on the PCA-transformed sampled data
kmeans_sampled = KMeans(n_clusters=5, random_state=0)
kmeans_sampled.fit(pca_sampled_data)
clusters_sampled = kmeans_sampled.labels_

# Adding cluster information to the sampled data for analysis
sampled_data['Cluster'] = clusters_sampled

# Plotting the explained variance for sampled data
plt.figure(figsize=(10, 6))
plt.bar(range(1, n_components_sampled + 1), explained_variance_sampled, alpha=0.7, align='center')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance by PCA Components (Sampled Data)')
plt.show()

n_components_sampled, explained_variance_sampled, sampled_data['Cluster'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Data Preprocessing

# Encoding categorical variables if necessary
label_encoders = {}
for column in data.select_dtypes(include=['object']).columns:
    if column != 'Attack_type':  # excluding the target variable
        le = LabelEncoder()
        data[column] = le.fit_transform(data[column])
        label_encoders[column] = le

# Splitting the dataset into features (X) and target (y)
X = data.drop('Attack_type', axis=1)
y = data['Attack_type']

# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Checking the shape of the training and testing sets
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier

# Model Selection: Random Forest Classifier
rf_classifier = RandomForestClassifier(random_state=0)

# Model Training
rf_classifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, accuracy_score

# Predictions on the Test Set
y_pred = rf_classifier.predict(X_test)

# Model Evaluation
accuracy = accuracy_score(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)

print("Accuracy:", accuracy)
print("Classification Report:\n", classification_rep)
```
